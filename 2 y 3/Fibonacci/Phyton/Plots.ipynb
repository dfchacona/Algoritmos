```python
import matplotlib.pyplot as plt
import numpy as np
from time import time
a=0
b=1

list= []

for j in range(2, 2000):
    n=j
    start_time = time()
    for i in range(2,n):

        ans=a+b
        a=b
        b=ans
        
    elapsed_time = time() - start_time
    list.append(elapsed_time)
returns =np.array(list)
plt.ylabel('Time (s)') # The data we generated is unitless, but don't forget units in general.
plt.xlabel('Term #(n)')
plt.plot(returns)
```
