```python

import matplotlib.pyplot as plt
import numpy as np
from time import time

listA=[]
for j in range(20):
    A = np.zeros((30, 30))
    A += np.random.normal(20, 0.02)
    B = np.zeros((30, 30))
    B += np.random.normal(10, 0.03)
    start_time = time()
    ans= A*B
    elapsed_time = time() - start_time
    listA.append(elapsed_time)
returns =np.array(listA)
print (returns)
plt.plot(returns)
```
